```python
import pandas as pd

import os
import pathlib
```

```python
root = os.getcwd()
src_root = os.path.join(root, 'csv')
src_root = pathlib.Path(src_root)
assert os.path.exists(src_root)
dst_dir = os.path.join(root, 'timeseries_by_attribute')

if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)

# Taking all other folders than ipynb checkpoints
dirs = [item for item in src_root.iterdir() if not 'ipynb' in item.stem]
```

```python
for directory in dirs:
    stem = directory.stem
    # used for fixing pet and skipping other
    if stem != 'pet':
        continue
    files = [item for item in directory.iterdir() if not 'ipynb' in item.stem]

    dfs = []

    for file in files:
        df = pd.read_csv(file)
        df['date'] = pd.to_datetime(df['date']) 
        df = df.set_index('date')
        dfs.append(df)
    result = pd.concat(dfs)
    result = result.sort_index()
    # ERA5-Land pet creates one extra day before time period starts
    result = result.loc["1961":"2023"]
    # Dropping off the hours
    result.index = result.index.date
    result.index =  result.index.set_names('date')
    dst_path = os.path.join(dst_dir, stem + '.csv')
    result.to_csv(dst_path)
```
